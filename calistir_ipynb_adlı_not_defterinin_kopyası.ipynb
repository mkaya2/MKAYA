```python
# -*- coding: utf-8 -*-
"""
Created on Mon Mar 17 13:26:43 2025

@author: asus
"""

import tkinter as tk
from tkinter import filedialog, messagebox
import numpy as np
import string
from collections import Counter
import itertools


def caesar_encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            shift_base = 65 if char.isupper() else 97
            encrypted_text += chr((ord(char) - shift_base + shift) % 26 + shift_base)
        else:
            encrypted_text += char
    return encrypted_text

def caesar_decrypt(text, shift):
    decrypted_text = ""
    for char in text:
        if char in string.ascii_letters:
            shift_amount = shift % 26
            if char.islower():
                new_char = chr(((ord(char) - ord('a') - shift_amount) % 26) + ord('a'))
            else:
                new_char = chr(((ord(char) - ord('A') - shift_amount) % 26) + ord('A'))
            decrypted_text += new_char
        else:
            decrypted_text += char
    return decrypted_text

def adjust_key(keyword, size=8):
    if len(keyword) < size:
        keyword = (keyword * (size // len(keyword) + 1))[:size]
    return keyword[:size]

def generate_key_order(keyword):
    adjusted_keyword = adjust_key(keyword)
    return np.argsort([ord(k) for k in adjusted_keyword])

def fill_matrix(text, size=8):
    blocks = []
    for i in range(0, len(text), size * size):
        block = text[i:i + size * size].ljust(size * size, 'X')
        matrix = np.array(list(block)).reshape(size, size)
        blocks.append(matrix)
    return blocks

def columnar_encrypt(text, keyword):
    key_order = generate_key_order(keyword)
    blocks = fill_matrix(text)
    encrypted_blocks = []
    for block in blocks:
        encrypted_matrix = block[:, key_order]
        encrypted_blocks.append(''.join(encrypted_matrix.flatten()))
    return ''.join(encrypted_blocks)

def decrypt_columnar_transposition(ciphertext, key):
    num_columns = len(key)
    num_rows = len(ciphertext) // num_columns
    if len(ciphertext) % num_columns != 0:
        num_rows += 1

    columns = [''] * num_columns
    index = 0

    for i in range(num_columns):
        col_length = num_rows
        if i < len(ciphertext) % num_columns:
            col_length += 1
        columns[i] = ciphertext[index:index + col_length]
        index += col_length

    sorted_columns = [''] * num_columns
    for i, column in enumerate(columns):
        sorted_columns[key[i] - 1] = column

    decrypted_text = ''
    for row in range(num_rows):
        for col in range(num_columns):
            if row < len(sorted_columns[col]):
                decrypted_text += sorted_columns[col][row]

    return decrypted_text

def brute_force_decrypt(input_file, output_file):
    with open(input_file, 'r') as file:
        ciphertext = file.read()

    key_length = 8
    possible_keys = list(itertools.permutations(range(1, key_length + 1)))

    for key in possible_keys:
        decrypted_text = decrypt_columnar_transposition(ciphertext, key)
        print(f"Attempt - Key: {key}")
        print(decrypted_text[:200])
        print("-" * 50)

def generate_vigenere_table():
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    table = []
    for i in range(len(alphabet)):
        table.append(list(alphabet[i:] + alphabet[:i]))
    return table

def vigenere_encrypt(text, key, table):
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    encrypted_text = ""
    key_index = 0
    for char in text:
        if char.upper() in alphabet:
            row = alphabet.index(key[key_index % len(key)].upper())
            col = alphabet.index(char.upper())
            encrypted_char = table[row][col]
            encrypted_text += encrypted_char.lower() if char.islower() else encrypted_char
            key_index += 1
        else:
            encrypted_text += char
    return encrypted_text

def decrypt_caesar_cipher(text):
    english_frequencies = {
        'e': 12.02, 't': 9.10, 'a': 8.12, 'o': 7.68, 'i': 7.31, 'n': 6.95,
        's': 6.28, 'r': 6.02, 'h': 5.92, 'd': 4.32, 'l': 3.98, 'u': 2.88,
        'c': 2.71, 'm': 2.61, 'f': 2.30, 'y': 2.11, 'p': 2.09, 'b': 1.49,
        'g': 1.43, 'v': 1.11, 'k': 0.69, 'x': 0.17, 'q': 0.11, 'j': 0.10,
        'z': 0.07, 'w': 0.06
    }

    text = text.lower()
    text = ''.join([char for char in text if char in string.ascii_lowercase])
    letter_counts = Counter(text)

    total_letters = sum(letter_counts.values())
    letter_frequencies = {char: (letter_counts[char] / total_letters) * 100 for char in letter_counts}

    most_common_letter = max(letter_frequencies, key=letter_frequencies.get)
    shift = (ord(most_common_letter) - ord('e')) % 26

    decrypted_text = ''
    for char in text:
        if char in string.ascii_lowercase:
            shifted_char = chr((ord(char) - shift - ord('a')) % 26 + ord('a'))
            decrypted_text += shifted_char
        else:
            decrypted_text += char

    return decrypted_text

def read_file(filename):
    with open(filename, "r", encoding="utf-8") as file:
        return file.read().strip()

def write_file(filename, text):
    with open(filename, "w", encoding="utf-8") as file:
        file.write(text)

def select_file():
    file_path = filedialog.askopenfilename()
    entry_file.delete(0, tk.END)
    entry_file.insert(0, file_path)

def select_output_file():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt", filetypes=[("Text files", "*.txt"), ("All files", "*.*")])
    entry_output_file.delete(0, tk.END)
    entry_output_file.insert(0, file_path)

def encrypt_text():
    file_path = entry_file.get()
    output_file = entry_output_file.get()
    method = encryption_method.get()
    if not file_path:
        messagebox.showwarning("Warning", "Please select a file.")
        return
    if not output_file:
        messagebox.showwarning("Warning", "Please select an output file.")
        return

    text = read_file(file_path)
    key = entry_key.get()
    encrypted_text = ""

    if method == "Caesar Cipher":
        shift = int(entry_shift.get()) if entry_shift.get().isdigit() else 3
        encrypted_text = caesar_encrypt(text, shift)
    elif method == "Columnar Transposition":
        if not key:
            messagebox.showwarning("Warning", "Please enter a keyword for Columnar Transposition.")
            return
        encrypted_text = columnar_encrypt(text, key)
    elif method == "Vigenère Cipher":
        if not key:
            messagebox.showwarning("Warning", "Please enter a keyword for Vigenère Cipher.")
            return
        vigenere_table = generate_vigenere_table()
        encrypted_text = vigenere_encrypt(text, key, vigenere_table)
    else:
        messagebox.showwarning("Warning", "Invalid encryption method.")
        return

    write_file(output_file, encrypted_text)
    messagebox.showinfo("Success", f"Encrypted file saved as {output_file}")

def decrypt_text():
    file_path = entry_file.get()
    if not file_path:
        messagebox.showwarning("Warning", "Please select a file.")
        return

    output_file = filedialog.asksaveasfilename(defaultextension=".txt", filetypes=[("Text files", "*.txt"), ("All files", "*.*")])
    if not output_file:
        return

    method = decryption_method.get()

    with open(file_path, "r", encoding="utf-8") as file:
        encrypted_text = file.read()

    if method == "Brute Force":
        brute_force_decrypt(file_path, output_file)
    elif method == "Frequency Analysis":
        decrypted_text = decrypt_caesar_cipher(encrypted_text)
        write_file(output_file, decrypted_text)
        messagebox.showinfo("Success", f"Decrypted file saved as {output_file}")
    else:
        messagebox.showwarning("Warning", "Please select a decryption method.")


root = tk.Tk()
root.title("Encryption & Decryption Tool")
root.geometry("1000x600")
root.config(bg="#2F2F2F")  # Set background to black

# Title Label
title_label = tk.Label(root, text="Encryption & Decryption Tool", font=("Helvetica", 16, "bold"), bg="#006400", fg="white")
title_label.pack(pady=10)

# main frame
main_frame = tk.Frame(root, bg="#2F2F2F")
main_frame.pack(pady=10, fill="both", expand=True)

# Left Frame
left_frame = tk.Frame(main_frame, bg="darkgray", width=400)
left_frame.pack(side="left", padx=20, pady=20, fill="y")

# Select File Section (Encryption)
frame_file = tk.Frame(left_frame, bg="darkgray")
frame_file.pack(pady=10, fill="x")
tk.Label(frame_file, text="Select File:", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
entry_file = tk.Entry(frame_file, width=40, font=("Helvetica", 10), bg="gray", fg="white")
entry_file.pack(side="left", padx=10)
tk.Button(frame_file, text="Browse", font=("Helvetica", 10), command=select_file, bg="#4CAF50", fg="white").pack(side="left")

# Select Output File Section (Encryption)
frame_output = tk.Frame(left_frame, bg="darkgray")
frame_output.pack(pady=10, fill="x")
tk.Label(frame_output, text="Select Output File:", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
entry_output_file = tk.Entry(frame_output, width=40, font=("Helvetica", 10), bg="gray", fg="white")
entry_output_file.pack(side="left", padx=10)
tk.Button(frame_output, text="Save As", font=("Helvetica", 10), command=select_output_file, bg="#4CAF50", fg="white").pack(side="left")

# Encryption Method Section
frame_encryption = tk.Frame(left_frame, bg="darkgray")
frame_encryption.pack(pady=10, fill="x")
tk.Label(frame_encryption, text="Encryption Method:", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
encryption_method = tk.StringVar(value="Caesar Cipher")
tk.OptionMenu(frame_encryption, encryption_method, "Caesar Cipher", "Columnar Transposition", "Vigenère Cipher").pack(side="left", padx=10)

# Key and Shift
frame_key_shift = tk.Frame(left_frame, bg="darkgray")
frame_key_shift.pack(pady=10, fill="x")
tk.Label(frame_key_shift, text="Key (if needed):", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
entry_key = tk.Entry(frame_key_shift, font=("Helvetica", 10), bg="gray", fg="white")
entry_key.pack(side="left", padx=10)
tk.Label(frame_key_shift, text="Shift (Caesar only):", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
entry_shift = tk.Entry(frame_key_shift, font=("Helvetica", 10), bg="gray", fg="white")
entry_shift.pack(side="left", padx=10)

# Encryption Button
frame_button_encrypt = tk.Frame(left_frame, bg="darkgray")
frame_button_encrypt.pack(pady=10, fill="x")
tk.Button(frame_button_encrypt, text="Encrypt", font=("Helvetica", 10), command=encrypt_text, bg="#4CAF50", fg="white").pack()

# Right Frame for Decryption Section
right_frame = tk.Frame(main_frame, bg="darkgray", width=400)
right_frame.pack(side="right", padx=20, pady=20, fill="y")

# Decryption Method Section
frame_decryption = tk.Frame(right_frame, bg="darkgray")
frame_decryption.pack(pady=10, fill="x")
tk.Label(frame_decryption, text="Decryption Method:", font=("Helvetica", 10), bg="#00008B", fg="white").pack(side="left", padx=10)
decryption_method = tk.StringVar(value="Frequency Analysis")
tk.OptionMenu(frame_decryption, decryption_method, "Brute Force", "Frequency Analysis").pack(side="left", padx=10)

# Decryption Button
frame_button_decrypt = tk.Frame(right_frame, bg="darkgray")
frame_button_decrypt.pack(pady=10, fill="x")
tk.Button(frame_button_decrypt, text="Decrypt", font=("Helvetica", 10), command=decrypt_text, bg="#4CAF50", fg="white").pack()

root.mainloop()
```
